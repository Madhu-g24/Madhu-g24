```python
def stamp_duty(house_price,first_time_buyer):
  if first_time_buyer:
    if house_price <= 300000:
      return print ("No stamp duty for first time buyers for house valuation under 300000")
    else:
      return house_price * 0.05
  else:
    if house_price <= 120000:
      return print ("No stamp duty for buyers for house valuation under 120000")
    else:
      return house_price * 0.05
```

```python
def fizzbuzz(n):
  """
  for multiples of both three and five, print fizzbuzz
  for multiples of three, print fizz
  for multiples of five, print buzz
  for others, print the number

  Keyword Arguments:
  n-- the number to test
  """
  if n % 3 == 0 and n % 5 == 0:
    print('Fizzbuzz')
  elif n % 3 == 0:
      print("Fizz")
  elif n % 5 == 0:
        print('buzz')
  else:
    print(n)
```

```python
n = 1
while n <= 15:
  print('n = {0:2}: '.format(n), end = '')
  fizzbuzz(n)
  n += 1 #n = n +1
```

```python
#identify missing values
missing = pd.isna(df).sum().sum()
print("Number of missing values in DataFrame: {0}".format(missing))
```

```python
#identify missing values using null
missing = pd.isnull(df).sum()
print("Number of missing values in DataFrame: {0}".format(missing))
```

```python
#delete missing value
deleted = df.dropna(how = 'any')
print(deleted)
```

```python
checking = pd.isnull(deleted).sum().sum()
print("no. of missing value now: {0}".format(checking))
```

```python
import numpy as np
num_filled = df.select_dtypes(include = np.number)
filled = num_filled.fillna(num_filled.mean())
print(filled)
```

```python
checking = pd.isnull(filled).sum().sum()
print("no. of missing value now: {0}".format(checking))
```

```python
#remove rows containing missing value in table
df = df.dropna(axis = 0)
print(df)
```

```python
#remove columns containing missing value in table
df = df.dropna(axis = 1)
print(df)
```
